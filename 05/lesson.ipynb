## 準備


### パッケージをインストール


```r
install.packages("kernlab")
library(kernlab)
```

### データの読み込み


```r
data(spam)
dim(spam)
```

```r
colnames(spam)
```

```r
summary(spam$money)
```

```r
set.seed(50)
tr.num = sample(4601, 2500)
tr.num
```

```r
spam.train = spam[tr.num, ]
spam.test <- spam[-tr.num, ]
```

## 分析してみよう


```r
set.seed(50)
spam.svm0 <- ksvm(type ~ ., kernel = "vanilladot", data = spam.train, cross = 3)
spam.svm0
```

```r
spam.pre0 <- predict(spam.svm0, spam.test[, -58])
spam.tab0 <- table(spam.test[, 58], spam.pre0)
spam.tab0
```

```r
spam.tab0[1, 2] + spam.tab0[2, 1]
```

# カーネル法


### さっきの SVM をガウシアンカーネル法を用いてやってみる


```r
set.seed(50)
spam.svm1 <- ksvm(type ~ ., data = spam.train, cross = 3)
spam.svm1
```

```r
spam.pre1 <- predict(spam.svm1, spam.test[, -58])
spam.tab1 <- table(spam.test[, 58], spam.pre1)
spam.tab1
```

```r
spam.tab1[1, 2] + spam.tab1[2, 1]
```

なぜか線形のときより悪くなってる
