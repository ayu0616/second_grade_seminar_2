```r
means = c()
for (i in 1:as.integer(length(UKgas) / 4)) {
    m <- mean(UKgas[(4 * i - 3):(4 * i)])
    means <- c(means, m)
}
```

```r
plot(ts(as.matrix(means), start = 1960, frequency = 1))
```

```r
means4 = c()
for (i in 1:length(means)) {
    for (j in 0:3) {
        means4 <- c(means4, means[i])
    }
}
ts.plot(ts(as.matrix(means4), start = c(1960, 1), frequency = 4), UKgas, lty = 2:1, col = c("red", "black"))
```
